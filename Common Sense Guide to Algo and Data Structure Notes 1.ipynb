```python
#å­—ç¬¦ä¸²æœç´¢ï¼Œå¦‚æœåœ¨haystackä¸­å‘ç°needleï¼Œè¿”å›True
def find_needle(needle, haystack):
    needle_index = 0
    haystack_index = 0
    for haystack_index in range(len(haystack)):
        if needle[needle_index] == haystack[haystack_index]:
            found_needle = True
            while needle_index < len(needle):
                if needle[needle_index] != haystack[haystack_index + needle_index]:
                    found_needle = False
                    needle_index = 0
                    break
                needle_index += 1
            if found_needle == True:
                return True
    return False
```

```python
#åˆ¤æ–­æ•°ç»„æ˜¯å¦æ˜¯å¦ä¸€ä¸ªæ•°ç»„çš„å­é›†ï¼ˆå¦‚æœä¸ç”¨hash,å¤æ‚åº¦ä¸ºO(nxm)),è¿™ä¸ªå¤æ‚åº¦ä¸ºO(n+m)
def isSubset(a1,a2):
    #åˆ¤æ–­å“ªä¸ªæ•°ç»„æ›´å°
    if len(a1) > len(a2):
        large_a = a1
        small_a = a2
    else:
        large_a = a2
        small_a = a1
    #æŠŠæ›´å¤§çš„æ•°ç»„å­˜å‚¨åˆ°hashtableä¸­
    ht = dict()
    for i in large_a:
        ht[i] = True
    #éå†å°æ•°ç»„ï¼Œcheckåœ¨ä¸åœ¨è¡¨ä¸­
    for i in small_a:
        if ht.get(i) == None:
            return False
    return True
```

```python
#è¿”å›ä¸¤ä¸ªæ•°ç»„çš„äº¤é›†ï¼ŒO(N)
def Overlap(a1,a2):
    ht = dict()
    overlap = []
    for i in a1:
        ht[i] = True
    for i in a2:
        if ht.get(i) != None:
            overlap.append(i)
    return overlap
```

```python
#æ£€æŸ¥ä¸€ä¸ªå­—ç¬¦ä¸²ä¸­çš„æ‹¬å·è¯­æ³•æ˜¯å¦æ­£ç¡®
#é”™è¯¯åŒ…æ‹¬ï¼š1.æœ‰å·¦æ‹¬å·ï¼Œæ²¡æœ‰å³ 2.æœ‰å³æ‹¬å·ï¼Œæ²¡æœ‰å·¦ 3.å·¦å³æ‹¬å·ä¸åŒ¹é…
#Define class of stack
class stack:
    def __init__(self):
        self.stack = []
    def append(self,a):
        self.stack.append(a)
    def pop(self):
        last_element = self.stack.pop()
        return last_element
    def len(self):
        return len(self.stack)
    def check(self):
        return self.stack
```

```python
def map_brackets(b):
    if b == "}":
        return "{"
    if b == "]":
        return "["
    if b == ")":
        return "("
```

```python
def check(s):
    left = stack()
    for i in s:
        if i not in ["{", "[", "(","}","]",")"]:
            continue
        elif i in ["{", "[", "("]:
            left.append(i)
        elif i in ["}","]",")"]:
            try:
                temp_pair = left.pop()
                if temp_pair == map_brackets(i):
                    continue
                else:
                    return "Error: Not Mapping "
            except:
                return "Error: No Left Bracket"
    print(left.check())
    if left.len() == 0:
        return "No Error"
    else:
        print(left.len())
        return "Error: Left Bracket Too Much"
```

```python
import queue

class PrintManager:
    def __init__(self):
        self.tasks = queue.Queue()
    def newtask(self,t):
        self.tasks.put(t)
    def printtask(self):
        t = self.tasks.get()
        print(t)
```

- å¹³å‡å¤æ‚åº¦ä¸ºO(N)

```python
class sort_array:
    def __init__(self,array):
        self.array = array
        self.len = len(array)
    
    #åˆ†åŒºï¼Œä½¿ä¸€ä¸ªæ•°ç»„ä¸­å°äºåŸºå‡†çš„æ•°éƒ½ä½äºåŸºå‡†çš„å·¦è¾¹ï¼Œå¤§äºåŸºå‡†çš„æ•°éƒ½ä½äºå³è¾¹
    def partition(self, left_pointer,right_pointer):
        pivot_index = right_pointer
        pivot = self.array[pivot_index]
        right_pointer -= 1
        # å‡ºé”™ï¼š<= å†™æˆäº† <
        while left_pointer <= right_pointer:
            if self.array[left_pointer] <= pivot:
                left_pointer += 1
            elif self.array[right_pointer] < pivot:
                self.array[left_pointer], self.array[right_pointer] = self.array[right_pointer], self.array[left_pointer]
            else:
                right_pointer -= 1
        self.array[left_pointer], self.array[pivot_index] = self.array[pivot_index], self.array[left_pointer] 
        #print(self.array)
        return left_pointer

    # åˆ†åŒºï¼Œé€’å½’
    def quicksort(self,left_pointer,right_pointer):
        if right_pointer <= left_pointer:
            return
        # å‡ºé”™ï¼šleft_pointerå†™æˆäº†0
        pivot = self.partition(left_pointer,right_pointer)
        self.quicksort(left_pointer,pivot -1)
        self.quicksort(pivot+1,right_pointer)

    def quickselect(self,kth_lowest_value, left_pointer, right_pointer):
        pivot = self.partition(left_pointer, right_pointer)
        if (kth_lowest_value - 1) == pivot:
            print(self.array[pivot])
            return (self.array[pivot])
        elif (kth_lowest_value - 1) < pivot:
            self.quickselect(kth_lowest_value,left_pointer,pivot-1)
        else:
            self.quickselect(kth_lowest_value,pivot+1,right_pointer)
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self,head):
        self.head = head
    def read(self,index):
        current_index = 0
        current_node = self.head
        while current_index < index:
            if current_node.next == None:
                return "Read: reached the tail of the LinkedList"
            current_node = current_node.next
            current_index += 1
        return current_node.data
    def index_of(self,value):
        current_index = 0
        current_node = self.head
        # çŠ¯é”™ï¼šå¾ªç¯æ¡ä»¶å†™æˆ while(current_node.next != None)
        while True:
            if current_node.data == value:
                return current_index
            if current_node.next == None:
                return "Searched for index: reached the tail of the LinkedList"
            current_node = current_node.next
            current_index += 1
    def insert_at_index(self,index,value):
        new_node = Node(value)
        current_index = 0
        current_node = self.head
        if index == 0:
            new_node.next = current_node
            self.head= new_node
            return
        if index > 0:
            while current_index < index - 1:
                if current_node.next == None:
                    return "Read: reached the tail of the LinkedList"
                current_node = current_node.next
                current_index += 1
            new_node.next = current_node.next
            current_node.next = new_node
            return
    def reverse(self):
        prev = self.head
        cur = self.head.next

        if cur==None:
            print("One Node, cannot reverse")
            return
        if cur.next == None:
            cur.next=prev
            self.head = cur
            return

        nex = cur.next
        
        while True:
            if nex.next == None:
                print("End")
                cur.next = prev
                nex.next = cur
                self.head = nex
                return
            cur.next = prev
            prev = cur
            cur = nex
            nex = nex.next
```

```python
class TreeNode:
    def __init__(self,val,left=None, right=None):
        self.value = val
        self.leftChild = left
        self.rightChild = right
    def isempty(self):
        return self.value == None
        
def insert(value, node):
    if value < node.value:
        if node.leftChild == None:
            node.leftChild = TreeNode(value)
        else:
            insert(value,node.leftChild)
    elif value < node.value:
        if node.rightChild == None:
            node.rightChild = TreeNode(value)
        else:
            insert(value,node.rightChild)
```

```python
# é”™è¯¯ä»£ç 
def delete(valueToDelete, node):
    if node == None:
        return None
    if node.value == valueToDelete:
        if (node.leftChild is None) and (node.rightChild is None):
            node = None
        elif node.leftChild is None:
            node = node.leftChild
        elif node.rightChild is None:
            node = node.rightChild
        else:
            houji = herit(node.rightChild)
            if not houji.rightChild is None:
                node = houji
                houji = houji.rightChild
            else:
                node = houji
                houji = None
    elif valueToDelete < node.value:
        delete(valueToDelete, node.leftChild)
    elif valueToDelete > node.value:
        delete(valueToDelete, node.rightChild)
```

```python
def herit(node):
    if node.leftChild == None:
        return node
    return herit(node.leftChild)
```

```python
def canPlaceFlowers(flowerbed, n: int) -> bool:
    maxnum = 0
    left_index = 0
    right_index = len(flowerbed) - 1
    if len(flowerbed) <= 2:
        if flowerbed.sum() == 0:
            return 1
        else:
            return 0
    else:
        while left_index <= right_index:
            if (flowerbed[left_index + 1] == 0) \
            and (flowerbed[max(left_index - 1, 0)] == 0) and (flowerbed[left_index] == 0):
                print("left ",left_index)
                flowerbed[left_index] = 1
                maxnum += 1
                if left_index == right_index:
                    break
            if (flowerbed[min(right_index + 1,len(flowerbed)-1)] == 0) \
            and (flowerbed[max(right_index - 1, 0)] == 0) and (flowerbed[right_index] == 0):
                print("right ", right_index)
                flowerbed[right_index] = 1
                maxnum += 1
            left_index += 1
            right_index -= 1
        return maxnum >= n     
```
